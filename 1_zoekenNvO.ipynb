



## 1. Zoeken naar een specifieke term, concordantie en export 



Eerst gaan we op zoek naar passages in de 577 digitale teksten waarin een bepaalde term voorkomt. In onderstaande code is het trefwoord een waarde van de variabele `search_term`(bv. “Baboe” of “geweer”). De aanhalingstekens geven aan dat deze waarde variabel is, je kunt deze zoekactie dus herhalen met steeds een ander trefwoord. 

Het resultaat van de zoekactie is een lijst met de metadata van alle documenten waarin de opgegeven term voorkomt, met daaronder  alle gevonden passages. Dit is een zogenaamde ‘concordantie’. In het Engels wordt dit een KWIC lijst genoemd, *keyword in context*. Met 'context' wordt bedoeld dat de passages bestaan uit een bepaald aantal woorden voor en na de term. Dit aantal kun je aangeven met de variabele `window`. Als je bv. de waarde 3 invoert, en je zoekt op “geweer”, dan kun je een zin terugkrijgen als ‘ik pak mijn geweer en schiet op’. In het onderstaande voorbeeld gebruiken we de term ‘baboe’ oftewel ‘wasvrouw’, die in veel van de memoires voorkomt. 

De metadata die van elk boek beschikbaar zijn, zijn de bibliografische gegevens. Daarnaast is aan elk boek voor het proces van digitaliseren een unieke numerieke code toegekend. Deze vind je terug in de bestandsnamen.

De functie 'showTitle()', in de module kitlvTdm, zoekt de volledige titels die bij deze documentcodes horen. Deze kunnen gevonden worden in het bestand 'metadata.csv'. Dat is een lijst met documentcodes, titels en andere metadata, die ook in de map met alle benodigde data staat, die je van te voren hebt moeten importeren. csv staat voor 'comma separated value', een bestandstype waarbij de waarden niet in aparte velden staan, maar door een comma worden gescheiden.  Dit is een makkelijk te 'lezen' format.

Om stap voor stap te kunnen begrijpen wat je de software vraagt te doen, vind je in elk veld lopende tekst met een uitleg van de afkortingen en de commandos. De uitleg kun je onderscheiden van de echte code doordat die wordt voorafgegaan door een hekje #  


```python
import os
import re
from os.path import join
from kitlvTdm import *
import pandas as pd

path = 'Corpus' 
search_term = 'baboe'
window = 10
max_books = 577
count = 0

for file in os.listdir( path ):
    if re.search( '[.]txt$' , file ) and count < max_books:
        count += 1
        book = open( join( path , file ), encoding = 'utf-8' )
        
        if re.search( search_term , book.read() , re.IGNORECASE):
            title = showTitle(file)
            concordances = concordance( join( path , file ) , search_term , window )
            
            if len(concordances) > 0:
                print( '\n\nConcordanties in:', title, file, '\n')
                
                for concordantie in concordances:
                    print(concordantie)
```

In de bovenstaande code worden de resultaten simpelweg getoond in dit Notebook. Je kunt uiteraard ook de resultaten exporteren naar een ander bestandstype, zodat je er ook in andere programma's mee kunt werken (Google sheets of Excel). Dan kun je kolommen toevoegen met je eigen commentaar of codes in de spreadsheet. 

Om dit te kunnen doe hebben we zogenaamde 'pandas' nodig, dat staat voor Python Data Analysis Library, of eenvoudiger Panel Data, een dataframe in Python taal, die je kunt exporteren.

een pandas bestand wordt dus herkend door Google sheets en Excel. 

De naam van dit tekstbestand wordt bepaald door de variabele 'outFile'.

In dit export bestand wordt ook het aantal treffers geteld, het totaal aantal treffers in het hele corpus, maar ook het aantal treffers per egodocument.

Ook hier volgt eerst de code, met daaronder per regel een uitleg. Omdat een heleboel regels hetzelfde zijn als in het vorige deel, worden alleen de nieuwe regels uitgelegd. 

Behalve code voor het exporteren van de resultaten in een dataframe, heb je ook code nodig om aan te geven waar de data naar toe moet. De code hieronder zorgt ervoor dat je tabel met zoekresultaten in dezelfde map terecht komt als waar de andere onderdelen die je gebruikt hebt staan.  


```python
#### path = 'Corpus'
import os
import re
from os.path import join
from kitlvTdm import *
import pandas as pd

searchTerm = 'baboe'
window = 10
outFile = 'concordantie'
df = pd.DataFrame(columns = ['count_within_document','file','title','fragment'])
for file in os.listdir( path ):
    if re.search( '[.]txt$' , file ):
        documentCount = 0 
        book = open( join( path , file ), encoding = 'utf-8' )
        
        if re.search( searchTerm , book.read() , re.IGNORECASE ):
            title = showTitle(file)
            concordances = concordance( join( path , file ) , searchTerm , window )
            
            for conc in concordances:
                documentCount += 1
                df = df.append({'count_within_document' : documentCount, 'file' : file,
                                'title' : title, 'match' : conc}, ignore_index = True)
                
df.to_excel(outFile + "concordantie" + "baboe" + ".xlsx")
print("Excel file saved as concordantie" + outFile + "concordantie" + "baboe" + ".xlsx" )


```
